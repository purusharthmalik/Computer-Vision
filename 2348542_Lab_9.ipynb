```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Plotting RGB images
def plot_image(img, title, og_im, og_title):    
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(og_im, cmap='gray')
    plt.title(og_title)
    plt.subplot(1, 2, 2)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.show()
```

### Loading the images

```python
im1 = cv2.imread("images/bird.png", cv2.IMREAD_GRAYSCALE)
im2 = cv2.imread("images/linus.jpeg", cv2.IMREAD_GRAYSCALE)
im3 = cv2.imread("images/wolverine.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(im1, cmap='gray')
plt.show()
plt.imshow(im2, cmap='gray')
plt.show()
plt.imshow(im3, cmap='gray')
plt.show()
```

### Point Detection

```python
point_filter = np.array([[1,1,1],
                         [1,-8,1],
                         [1,1,1]])
threshold = 220

# Filtering the image
point_img = cv2.filter2D(src=im1, ddepth=-1, kernel=point_filter)

# Setting the threshold
for row in range(point_img.shape[0]):
    for col in range(point_img.shape[1]):
        if point_img[row][col] < threshold:
            point_img[row][col] = 0

plot_image(point_img, "Point Detection", im1, "Original Image")
```

```python
# Filtering the image
point_img = cv2.filter2D(src=im2, ddepth=-1, kernel=point_filter)

# Setting the threshold
for row in range(point_img.shape[0]):
    for col in range(point_img.shape[1]):
        if point_img[row][col] < threshold:
            point_img[row][col] = 0

plot_image(point_img, "Point Detection", im2, "Original Image")
```

```python
# Filtering the image
point_img = cv2.filter2D(src=im3, ddepth=-1, kernel=point_filter)

# Setting the threshold
for row in range(point_img.shape[0]):
    for col in range(point_img.shape[1]):
        if point_img[row][col] < threshold:
            point_img[row][col] = 0

plot_image(point_img, "Point Detection", im3, "Original Image")
```

### Line Detection

```python
# Horizontal Line Detection
hline_filter = np.array([[-1,-1,-1],
                         [2,2,2],
                         [-1,-1,-1]])

# Vertical Line Detection
vline_filter = np.array([[-1,2,-1],
                         [-1,2,-1],
                         [-1,2,-1]])

# Filtering the image
hor_img = cv2.filter2D(src=im1, ddepth=-1, kernel=hline_filter)
ver_img = cv2.filter2D(src=im1, ddepth=-1, kernel=vline_filter)

plot_image(hor_img, "Horizontal Line Detection", im1, "Original Image")
plot_image(ver_img, "Vertical Line Detection", im1, "Original Image")
```

```python
# Filtering the image
hor_img = cv2.filter2D(src=im2, ddepth=-1, kernel=hline_filter)
ver_img = cv2.filter2D(src=im2, ddepth=-1, kernel=vline_filter)

plot_image(hor_img, "Horizontal Line Detection", im2, "Original Image")
plot_image(ver_img, "Vertical Line Detection", im2, "Original Image")
```

```python
# Filtering the image
hor_img = cv2.filter2D(src=im3, ddepth=-1, kernel=hline_filter)
ver_img = cv2.filter2D(src=im3, ddepth=-1, kernel=vline_filter)

plot_image(hor_img, "Horizontal Line Detection", im3, "Original Image")
plot_image(ver_img, "Vertical Line Detection", im3, "Original Image")
```

### Slant Line Detection

```python
# Positive Slant Line Filter
pos_slant_filter = np.array([[-1, -1, 2], 
                             [-1, 2, -1], 
                             [2, -1, -1]])

# Negative Slant Line Filter
neg_slant_filter = np.array([[2,-1,-1],
                             [-1,2,-1],
                             [-1,-1,2]])

# Filtering the image
pos_img = cv2.filter2D(src=im1, ddepth=-1, kernel=pos_slant_filter)
neg_img = cv2.filter2D(src=im1, ddepth=-1, kernel=neg_slant_filter)

plot_image(pos_img, "+45 Slanted Image", im1, "Original Image")
plot_image(neg_img, "-45 Slanted Image", im1, "Original Image")
```

```python
# Filtering the image
pos_img = cv2.filter2D(src=im2, ddepth=-1, kernel=pos_slant_filter)
neg_img = cv2.filter2D(src=im2, ddepth=-1, kernel=neg_slant_filter)

plot_image(pos_img, "+45 Slanted Image", im2, "Original Image")
plot_image(neg_img, "-45 Slanted Image", im2, "Original Image")
```

```python
# Filtering the image
pos_img = cv2.filter2D(src=im3, ddepth=-1, kernel=pos_slant_filter)
neg_img = cv2.filter2D(src=im3, ddepth=-1, kernel=neg_slant_filter)

plot_image(pos_img, "+45 Slanted Image", im3, "Original Image")
plot_image(neg_img, "-45 Slanted Image", im3, "Original Image")
```
