```python
#importing libraries
import cv2
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import PIL
```

## Reading and viewing the image with different libraries

```python
#cv2
im = cv2.imread("images/bird.png")
rgb_im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_im)
plt.title("Using OpenCV")
plt.show()
```

```python
#imageio
im = imageio.imread("images/bird.png")
plt.imshow(im)
plt.title("Using ImageIO")
plt.show()
```

```python
#matplotlib
im = plt.imread("images/bird.png")
plt.imshow(im)
plt.title("Using Matplotlib")
plt.show()
```

```python
#PIL
im = PIL.Image.open("images/bird.png")
plt.imshow(im)
plt.title("Using PIL")
plt.show()
```

## Writing images while changing its format

```python
#imageio
im = imageio.imread("images/bird.png")
imageio.imsave("bird_new.jpeg", im)
```

```python
#checking if the file was saved
im = PIL.Image.open("bird_new.jpeg")
im
```
