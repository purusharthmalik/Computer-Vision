```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Plotting RGB images
def plot_image(img, title, og_im = cv2.cvtColor(cv2.imread('bird_new.jpeg'), cv2.COLOR_BGR2RGB), og_title="Original Image"):    
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(og_im)
    plt.title(og_title)
    plt.subplot(1, 2, 2)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.show()
```

```python
image = cv2.cvtColor(cv2.imread('bird_new.jpeg'), cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### Demonstrate the following edge detection methods  and interpret the results
- First Order Derivative Methods

We will go with `Sobel Edge Detector` for the first order derivative,

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
sobel_x = cv2.Sobel(image_gray, -1, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image_gray, -1, 0, 1, ksize=5)

plot_image(sobel_x, "Sobel Edge Detection - Horizontal (First Order Derivative)")
plot_image(sobel_y, "Sobel Edge Detection - Vertical (First Order Derivative)")
```

- Second Order Derivative Methods

For this, we will use a `Laplacian Based Edge Detector`.

```python
from PIL import Image, ImageFilter

img = Image.open("images/bird.png")
laplacian = img.convert("L").filter(ImageFilter.FIND_EDGES)

plot_image(laplacian, "Laplacian Based Edge Detection (Second Order Derivative)")
```

- Optimum Edge Detection Method

```python
canny = cv2.Canny(image, 100, 200)
plot_image(canny, title="Canny Edge Detection (Optimum Edge Detection)")
```

### Results

Clearly, `Canny Edge Detection` provides the best results as it is robust to noise and produces thin, well-defined edges.
