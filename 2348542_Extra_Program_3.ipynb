## Use the following interpolation techniques in sampling:

1. Bilinear
2. Bicubic
3. Nearest Neighbor
4. LANCZOS4

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

im = cv2.imread('images/bird.png')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

plt.imshow(im)
plt.title('Original Image')
plt.show()
```

### Bilinear

```python
im_bilinear = cv2.resize(im, (128, 128), interpolation=cv2.INTER_LINEAR)
plt.imshow(im_bilinear)
plt.title('Bilinear Sampling')
plt.show()
```

### Bicubic

```python
im_bicubic = cv2.resize(im, (128, 128), interpolation=cv2.INTER_CUBIC)
plt.imshow(im_bicubic)
plt.title('Bicubic Sampling')
plt.show()
```

### Nearest Neighbor

```python
im_nn = cv2.resize(im, (128, 128), interpolation=cv2.INTER_NEAREST)
plt.imshow(im_nn)
plt.title('Nearest Neighbor Sampling')
plt.show()
```

### LANCZOS4

```python
im_lan = cv2.resize(im, (128, 128), interpolation=cv2.INTER_LANCZOS4)
plt.imshow(im_lan)
plt.title('LANCZOS4 Sampling')
plt.show()
```

### Comparing all the techniques

```python
fig, ax = plt.subplots(1, 4, figsize=(16, 8))

titles = ['Bilinear', 'Bicubic', 'Nearest Neighbor', 'LANCZOS4']

for idx, tech in zip(range(4), [im_bilinear, im_bicubic, im_nn, im_lan]):
    ax[idx].imshow(tech)
    ax[idx].title.set_text(titles[idx])
    ax[idx].axis('off')
```
